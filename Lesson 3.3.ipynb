```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pokemon = pd.read_csv('pokemon.csv')
```

Question 1:

Histogram - Not a normal distribution so we can not really get an idea of the distribution.
Box Plot - Gives us a good look at the outliers in this data set. 
Count Plot - Variables have two wide of a range to make the chart clear to read. 
Boxen Plot - Similar to the Box Plot it gives us a great look at the distribution of the variables. 

```python
hp= pokemon['HP']
sns.distplot(hp,color='red')
plt.title('Histogram of HP')
plt.show()

sns.boxplot(hp)
plt.title('BoxPlot of HP')
plt.show()
sns.countplot(x=hp)
plt.title('Count of HP')
plt.show()

sns.boxenplot(x=hp)
plt.title('BoxenPlot of HP')
plt.show()
```

Question 2: 
- Scatter Plot - Gives us a good look at the distribution of HP and Defense and shows that we do not have a direct correlation between the two. 
- Hex Plot - Like the scatter plot shows us the distribution of the two variable but does not do as good of a job of show us outliers.
- Line Plot - Really does not provied much insight on these particular variables. 

```python
g = sns.lmplot(y='HP', 
               x='Defense',
               data=pokemon, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("HP")
g.set_xlabels("Defense")
plt.title('Scatterplot of Defense and HP')
plt.show()

sns.jointplot(x= 'HP', y= 'Defense', kind="hex", data = pokemon, color="#4CB391")
plt.title('Hexplot of Defense and HP')
plt.show()
## Trouble centering title

sns.lineplot(x="HP", y="Defense",
             data=pokemon)
plt.title('Line Plot of Defense and HP')
plt.show()
```

Question 3: 
- Barplot  - Does not show us the low end of values because it starts at 0 . 
- Point Plot - Achieves what the Barplot does, but with the added ebnifit of also showing the low values. 
- Box Plot - does a good job of showing the distributions across the Types as well as identifying outliers. 
- Violin Plot - Similarly to the Box Plot helps identiy the outliers, but does a slighty better job showing the where each Type is most likely to fall. 
- Swarm Plot - very similar to the violin plot with less of an visual emphasis on where the value is most likely. 
- Replot - Very similar to the swarm plot, but realies on hues to show you where the most likely values is. Makes it harder to visually interpret.

```python
### Hit Points by Type
g = sns.catplot(x="Type 1", y="HP", data=pokemon, 
                 kind="bar", palette="pastel",
                 height=8, aspect= 2 / 1)
g.despine(left=True)
g.set_ylabels("HP")
g.set_xlabels("Type")
plt.title('Barplot: HP by Type')
plt.show()

g = sns.catplot(x="Type 1", y="HP", data=pokemon, 
                 kind="point", palette="pastel",
                 height=6, aspect= 2 / 1,ci=95)
g.despine(left=True)
g.set_ylabels("HP")
g.set_xlabels("Type")
plt.title('Point Plot: HP by Type')
plt.show()


g = sns.catplot(x="Type 1", y="HP", data=pokemon, 
                 kind="box", palette="pastel",
                 height=6, aspect= 2 / 1,ci=95)
g.despine(left=True)
g.set_ylabels("HP")
g.set_xlabels("Type")
plt.title('Box Plot: HP by Type')
plt.show()
```

```python
plt.rcParams['figure.figsize']=(12,12)
sns.violinplot(x="Type 1", y="HP",
               data=pokemon)
plt.show()


sns.swarmplot(x="Type 1", y="HP",
              data=pokemon)
g.set_ylabels("HP")
g.set_xlabels("Type")
plt.title('Swarm Plot: HP by Type')
plt.show()

```

```python
sns.relplot(x="Type 1", y="HP",
            sizes=(40, 400), alpha=.5,
            height=10, data=pokemon)
plt.title('Replot: HP by Type')
plt.show()
```
